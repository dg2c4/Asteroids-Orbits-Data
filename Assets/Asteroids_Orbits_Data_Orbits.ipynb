```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV
df = pd.read_csv('/content/orbits.csv', delimiter=';')

# Mostrar las primeras filas del DataFrame
print(df.head())

# Información general del DataFrame
print(df.info())

# Estadísticas descriptivas
print(df.describe())

# Limpieza de datos: Eliminar filas con valores nulos
df = df.dropna()
```

```python
# Recargar el archivo con el delimitador correcto
df = pd.read_csv('/content/orbits.csv', delimiter=';')

# Mostrar las primeras filas para entender la estructura del archivo
df.head()
```

# Distribución de Asteroides por Clasificación: Diagrama De Barras.

```python
# 1. Distribución de los asteroides por clasificación
plt.figure(figsize=(17, 6))
sns.countplot(data=df, x='Object Classification', order=df['Object Classification'].value_counts().index, color="green")
plt.title('Distribución de Asteroides por Clasificación')
plt.xlabel('Clasificación')
plt.ylabel('Cantidad')
plt.show()
```

# Correlación entre la excentricidad de la órbita y la distancia mínima de intersección: Gráfico de dispersión.

```python
# 2. Correlación entre excentricidad y distancia mínima de intersección
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Orbit Eccentricity', y='Minimum Orbit Intersection Distance (AU)', color="blue")
plt.title('Correlación entre Excentricidad y Distancia Mínima de Intersección')
plt.xlabel('Excentricidad de la Órbita')
plt.ylabel('Distancia Mínima de Intersección (AU)')
plt.show()
```

# Distribución de las magnitudes absolutas de los asteroides: Histograma.

```python
# 3. Distribución de las magnitudes absolutas de los asteroides
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Asteroid Magnitude', bins=30, kde=True, color="blue")
plt.title('Distribución de las Magnitudes Absolutas de los Asteroides')
plt.xlabel('Magnitud Absoluta')
plt.ylabel('Frecuencia')
plt.show()
```

# Distribución de las distancias mínimas de intersección con la Tierra: Histograma.

```python
# 4. Distribución de las distancias mínimas de intersección con la Tierra
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Minimum Orbit Intersection Distance (AU)', bins=30, kde=True, color="purple")
plt.title('Distribución de las Distancias Mínimas de Intersección con la Tierra')
plt.xlabel('Distancia Mínima de Intersección (AU)')
plt.ylabel('Frecuencia')
plt.show()
```

# Relación entre la magnitud absoluta y la distancia mínima de intersección: Grafico de dispersión.

```python
# 5. Relación entre la magnitud absoluta y la distancia mínima de intersección
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Asteroid Magnitude', y='Minimum Orbit Intersection Distance (AU)', color="orange")
plt.title('Relación entre Magnitud Absoluta y Distancia Mínima de Intersección')
plt.xlabel('Magnitud Absoluta')
plt.ylabel('Distancia Mínima de Intersección (AU)')
plt.show()
```

# Lista de clasificaciones de asteroides y sus excentricidades: Grafico de caja- violín.

```python
# 6. Diagrama de caja para la excentricidad por clasificación de asteroides
plt.figure(figsize=(17, 6))
sns.boxplot(data=df, x='Object Classification', y='Orbit Eccentricity', color="skyblue")
plt.title('Excentricidad de la Órbita por Clasificación de Asteroides')
plt.xlabel('Clasificación')
plt.ylabel('Excentricidad de la Órbita')
plt.show()
```

# Relación entre el eje orbital y la distancia mínima de intersección: Grafico de dispersión.

```python
# 7. Diagrama de dispersión entre el eje orbital y la distancia mínima de intersección
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Orbit Axis (AU)', y='Minimum Orbit Intersection Distance (AU)', color="blue")
plt.title('Relación entre el Eje Orbital y la Distancia Mínima de Intersección')
plt.xlabel('Eje Orbital (AU)')
plt.ylabel('Distancia Mínima de Intersección (AU)')
plt.show()
```

#  Distribución de clasificaciones de asteroides. Gráfico Pastel.

```python
# Crear el diagrama de pastel
plt.figure(figsize=(8, 14))
df['Object Classification'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribución de Clasificaciones de Asteroides')
plt.ylabel('')

# Agregar leyendas con un espacio o un cuadro de conversiones
plt.legend(title="Clasificaciones", loc="upper right", bbox_to_anchor=(0, 1))

# Mostrar el gráfico
plt.show()
```
